# Подготовка данных

прицепим последнее действие к мейлу и уберем конкретные даты из данных и меил

```python
train_add.sort_values(by=['email', 'calday'], inplace=True)
train_last_calday=pd.DataFrame(train_add.drop_duplicates(subset='email', keep='last'))
```

```python
joined_data=pd.merge(train, train_last_calday, on='email', how='left')
joined_data.head(2)
```

```python
joined_data['action']=joined_data['action'].apply(lambda x: 1 if x=='CREATE' else 0)
joined_data['trial']=joined_data['trial'].apply(lambda x: 1 if x=='True' else 0)
```

```python
print(len(train), len(joined_data))
joined_data.describe(include='all')
```

кодируем текстовые переменные

```python
y=joined_data['label']
X=joined_data.drop(['label', 'cut_date', 'days_to_end',  'first_date', 'last_date', 'email', 'calday'], axis=1)
```

```python
from sklearn.model_selection import train_test_split,  GridSearchCV

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_val=scaler.transform(X_val)
```

```python
from sklearn.metrics import roc_auc_score
```

# Наивный

ну стоило попробовать)

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf_NB = GaussianNB()
clf_NB = clf_NB.fit(X_train ,y_train) 
print('train score: ', roc_auc_score( y_train, clf_NB.predict_proba(X_train)[:,1]),'\n',
      'val score: ', roc_auc_score( y_val, clf_NB.predict_proba(X_val)[:,1]))
```

# Обучение дерева

это не kaggle, так что не будем задротить на мелкой сетке

```python
from sklearn.tree import DecisionTreeClassifier

clf_tree = DecisionTreeClassifier(criterion='entropy', random_state=42)
params = {'max_depth': range(40, 100, 20)}
search_tree = GridSearchCV(clf_tree, params, scoring='roc_auc')

search_tree.fit(X_train, y_train)
```

```python
print(search_tree.best_params_, '\n',
      'refit_time_:', search_tree.refit_time_, '\n',
      'train score: ', search_tree.best_score_ ,'\n',
      'val score: ', roc_auc_score(y_val, search_tree.best_estimator_.predict_proba(X_val)[:,1]))
```
